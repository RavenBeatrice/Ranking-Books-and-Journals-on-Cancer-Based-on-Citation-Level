```python
import csv
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_word_cloud(csv_file, column_name):
    words = []
    with open(csv_file, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            words.extend(row[column_name].split())

    # Kelimeleri birleştirerek metin oluşturma
    text = ' '.join(words)

    # WordCloud nesnesi oluşturma
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    # Kelime bulutunu çizdirme
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Örnek kullanım
csv_file = "C:\\Users\\yunus\\Desktop\\Veri Manipülasyonu\\ARTICLE-INFO\\articles_df.csv"  # CSV dosya adını buraya yazın
column_name = 'Source Title'  # İlgili sütunun adını buraya yazın

generate_word_cloud(csv_file, column_name)
```
