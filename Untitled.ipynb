```python
import pandas as pd
import glob
```

```python
# Dosya yolu ve adını belirtin
file_location = "C:\\Users\\yunus\\Desktop\\Veri Manipülasyonu\\ARTICLE-INFO\\*"
# dosya_yolu değişkenine '*' yerleştirerek tüm xls dosyalarını seçiyoruz

# Tüm dosyaları bir listeye okuyun
all_files = glob.glob(file_location)

# Verileri depolamak için boş bir pandas DataFrame oluşturun
articles_df = pd.DataFrame()

# Tüm dosyaları tek tek okuyun ve birleştirin
for file in all_files:
    # xls dosyalarını okuyoruz
    tables = pd.read_excel(file)
    # birleştirme işlemi
    articles_df = pd.concat([articles_df, tables])

# Birleştirilen tabloyu csv dosyasına kaydedin
articles_df.to_csv("C:\\Users\\yunus\\Desktop\\Veri Manipülasyonu\\ARTICLE-INFO\\articles_df.csv")
```
